```python
from pathlib import Path
import pandas as pd

dir = Path("/Net/Groups/BGI/people/ecathain/TRENDY_Emulator_Scripts/NewModel/pipeline/3.benchmark/ilamb/benchmarks/stabilised_scenarios")

dfs = []

for S in ['S0', 'S1', 'S2', 'S3']:
    p = dir / S / "_build/scores.csv"
    df_s = pd.read_csv(p)

    df_s = df_s.set_index("Variables")
    df_s = df_s[["32_year"]].rename(columns={"32_year": S})

    dfs.append(df_s)

df_combined = pd.concat(dfs, axis=1)

# Add the average column
# Add the average column using only numeric columns
df_combined["Mean"] = df_combined.mean(axis=1, numeric_only=True)

df_combined.to_csv("per_scenario_scores.csv")
print(df_combined.head())
```
