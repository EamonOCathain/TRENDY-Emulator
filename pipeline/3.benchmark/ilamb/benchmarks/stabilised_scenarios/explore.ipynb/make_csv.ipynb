```python
import pandas as pd
S0 = pd.read_csv("/Net/Groups/BGI/people/ecathain/TRENDY_Emulator_Scripts/NewModel/pipeline/3.benchmark/ilamb/benchmarks/stabilised_scenarios/S0/_build/scalar_database.csv")
S1 = pd.read_csv("/Net/Groups/BGI/people/ecathain/TRENDY_Emulator_Scripts/NewModel/pipeline/3.benchmark/ilamb/benchmarks/stabilised_scenarios/S1/_build/scalar_database.csv")
S2 = pd.read_csv("/Net/Groups/BGI/people/ecathain/TRENDY_Emulator_Scripts/NewModel/pipeline/3.benchmark/ilamb/benchmarks/stabilised_scenarios/S2/_build/scalar_database.csv")
S3 = pd.read_csv("/Net/Groups/BGI/people/ecathain/TRENDY_Emulator_Scripts/NewModel/pipeline/3.benchmark/ilamb/benchmarks/stabilised_scenarios/S3/_build/scalar_database.csv")
scens = [S0, S1, S2, S3]

print(S0.head())
```

```python
S = S0.copy()
```

```python
final = {}
counter = 0
for S in scens:
    # Filter to Region == "Test"
    S_test = S[S["Region"] == "global"]

    # Keep only the columns we need
    S_small = S_test[["Variable", "ScalarName", "Data"]]

    # Pivot into your desired shape
    S_wide = S_small.pivot_table(
        index="Variable",
        columns="ScalarName",
        values="Data",
        aggfunc="first"     # each cell should only have one value
    )

    # Optional: order columns the way ILAMB normally outputs them
    desired_order = [
        "Bias Score",
        "RMSE Score",
        "Spatial Distribution Score",
        "Seasonal Cycle Score",
        "Overall Score",
    ]
    S_wide = S_wide.reindex(columns=desired_order)

    final[f"S{counter}"] = S_wide
    counter += 1
    

```
