```python
def write_variable(var: str,
                            da: xr.DataArray,
                            out_root: Path,
                            model, 
                            start_year: int = 1901,
                            end_year: int = 2023,):
    """
    Convert VISIT-UT state variables to monthly resolution and save.

    Parameters
    ----------
    var : str
        Variable name (e.g., 'cVeg', 'cLitter', 'mrso', 'lai', 'cTotal')
    da : xr.DataArray
        Input DataArray already masked (e.g., masked_visit_ut[var])
    out_root : Path
        Output directory root where VISIT-UT subdirectory will be written.
    start_year : int
    end_year : int
    """

    monthly_vars = ['mrso', 'lai', 'cTotal']
    out_dir = out_root / model
    out_dir.mkdir(parents=True, exist_ok=True)
    
    # Mask
    da = da.where(tvt_mask.isin([0, 1, 2]))

    # ------------------------
    # Case 1: Already monthly
    # ------------------------
    if var in monthly_vars or len(da.time) == (end_year - start_year + 1) * 12:
        out_path = out_dir / f"{model}_S3_{var}.nc"
        if out_path.exists():
            out_path.unlink()
        da.to_netcdf(out_path)
        print(f"[monthly] Saved {var} → {out_path}")
        return

    # --------------------------------------
    # Case 2: Annual → expand to monthly
    # --------------------------------------
    n_years = da.sizes["time"]
    rep_idx = np.repeat(np.arange(n_years), 12)
    da_m = da.isel(time=rep_idx)

    # Expect exactly (end_year - start_year + 1) years
    expected_years = end_year - start_year + 1
    if n_years != expected_years:
        raise ValueError(
            f"Expected {expected_years} years for {var}, but got {n_years}."
        )

    # Build noleap monthly timestamps
    years = np.repeat(np.arange(start_year, end_year + 1), 12)
    months = np.tile(np.arange(1, 13), expected_years)

    new_time = [
        cftime.DatetimeNoLeap(int(y), int(m), 1)
        for y, m in zip(years, months)
    ]

    da_m = da_m.assign_coords(time=("time", new_time))
    da_m.time.encoding.update({
        "units": "days since 1901-01-01 00:00:00",
        "calendar": "noleap",
    })

    out_path = out_dir / f"{model}_S3_{var}.nc"
    if out_path.exists():
        out_path.unlink()
    da_m.to_netcdf(out_path)

    print(f"[annual→monthly] Saved {var} → {out_path}")
```

```python
out_dir = Path("/Net/Groups/BGI/people/ecathain/TRENDY_Emulator_Scripts/NewModel/pipeline/3.benchmark/ilamb/benchmarks/transfer_learning_from_1982/AllTrendyModels")

for var, da in visit_ut_das.items():
    write_variable(var, da, out_dir, "VISIT-UT")

for var, da in visit_das.items():
    write_variable(var, da, out_dir, "VISIT")

for var, da in elm_das.items():
    write_variable(var, da, out_dir, "ELM")
```
