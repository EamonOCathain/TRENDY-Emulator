```python
import pandas   

csv = "/Net/Groups/BGI/people/ecathain/TRENDY_Emulator_Scripts/NewModel/pipeline/3.benchmark/ilamb/benchmarks/no_rolling_mean/_build/scalar_database.csv"

df = pandas.read_csv(csv)

print(df)
```

```python
df = df[(df['ScalarName'] == "Spatial Distribution Score") | ((df['ScalarName'] == "Overall Score") & (df['Region']=="test"))]
df.drop(columns=['Source','Section','Region','ScalarType','AnalysisType', 'Units', 'Weight'], inplace=True)
print(df)
```

```python
import pandas as pd

df_wide = (
    df
    .pivot_table(
        index="Variable",
        columns=["Model", "ScalarName"],
        values="Data"
    )
    .sort_index(axis=1, level=[0, 1])
)
```

```python
import pandas as pd

# 1. Pivot to wide: rows = Variable, cols = (Model, ScalarName)
wide = df.pivot_table(
    index="Variable",
    columns=["Model", "ScalarName"],
    values="Data"
)

# 2. Extract the two model blocks
base = wide["Base-Emulator_no_carry"]
no_roll = wide["no_rolling_means_no_carry"]

# 3. Compute deltas: no_rolling - base
delta = no_roll - base   # DataFrame with same columns as base/no_roll

# 4. Keep Spatial on the left, then Overall, and rename headers
delta = delta[["Spatial Distribution Score", "Overall Score"]]
delta.columns = ["Delta Spatial Distribution", "Delta Overall Score"]

# `delta` is now your final table:
# index: Variable
# columns: Delta Spatial Distribution, Delta Overall Score
print(delta)

delta.to_csv("delta_scores.csv")
```
