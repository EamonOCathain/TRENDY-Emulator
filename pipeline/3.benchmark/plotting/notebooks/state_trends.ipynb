```python

def xr_linear_trend_years(da: xr.DataArray, time_dim: str = "time") -> xr.DataArray:
    """
    Linear trend per grid cell along <time_dim>, in units per year.

    Assumes:
      - <time_dim> is datetime-like (cftime or numpy datetime)
      - calendar is effectively noleap, but only year + dayofyear are used.
    """
    time = da[time_dim]

    # fractional year: year + (doy-1)/365
    years = time.dt.year.astype("float64") + (
        time.dt.dayofyear.astype("float64") - 1.0
    ) / 365.0
    t = xr.DataArray(years, dims=time_dim)

    t_centered = t - t.mean(dim=time_dim)
    da_centered = da - da.mean(dim=time_dim)

    slope = (da_centered * t_centered).sum(dim=time_dim) / (t_centered**2).sum(dim=time_dim)

    if da.name:
        slope = slope.rename(da.name + "_trend")
    return slope

def trend_lai_or_mrso(da: xr.DataArray, time_dim: str = "time") -> xr.DataArray:
    """
    Remove monthly climatology, then compute linear trend per year.
    """
    clim = da.groupby(f"{time_dim}.month").mean(dim=time_dim)
    anomalies = da.groupby(f"{time_dim}.month") - clim
    return xr_linear_trend_years(anomalies, time_dim=time_dim)

def annual_mean_trend(da: xr.DataArray, time_dim: str = "time") -> xr.DataArray:
    """
    Take annual mean first, then compute linear trend per year.
    """
    # groupby year â†’ da_ann has dimension 'year'
    da_ann = da.groupby(f"{time_dim}.year").mean(dim=time_dim)

    # ensure 'year' is a coordinate (it already is, but cast to float for safety)
    da_ann = da_ann.assign_coords(year=da_ann["year"].astype("float64"))

    return xr_linear_trend_over_year_coord(da_ann, year_dim="year")

def xr_linear_trend_over_year_coord(da: xr.DataArray, year_dim: str = "year") -> xr.DataArray:
    """
    Linear trend per grid cell along a numeric 'year' coordinate.
    da[year_dim] must be 1D and numeric (e.g. 1901, 1902, ...).
    """
    years = da[year_dim].astype("float64")
    t = xr.DataArray(years, dims=year_dim)

    t_centered = t - t.mean(dim=year_dim)
    da_centered = da - da.mean(dim=year_dim)

    slope = (da_centered * t_centered).sum(dim=year_dim) / (t_centered**2).sum(dim=year_dim)

    if da.name:
        slope = slope.rename(da.name + "_trend")
    return slope
```
